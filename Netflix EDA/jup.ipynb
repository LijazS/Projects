importing libraries

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

loading csv


```python
df = pd.read_csv('C:/Users/lijaz/Desktop/DATA/netflix/netflix1.csv')
```

selecting all movies and grouping by director

```python
bf = df[df['type']=='Movie']
bf = bf.groupby(['director']).size().reset_index(name='count')
bf.head(20)
```

Removing not given directors

```python
bf = bf[bf['director'] != 'Not Given']
```

selecting all directors with more than one movie

```python
cf = bf[bf['count'] >10]
cf.head(10)
```

```python
cf.shape
```

SORTING BY COUNT IN DESCENDING ORDER

```python
x = cf.sort_values(by=['count'],ascending=False)
x.head()
```

PLOTTING A BAR CHART with director and no. of their movies

```python


labels = x['director']
values = x['count']

plt.figure(figsize=(10, 6))

plt.barh(labels,values,color = 'skyblue')

plt.xticks(range(0,25,2))


plt.show()

```

Grouping All data by release year and the count of release year

```python
df = pd.read_csv('C:/Users/lijaz/Desktop/DATA/netflix/netflix1.csv')

bf = df.groupby(['release_year']).size().reset_index(name='count')
```

Selecting movies release after and before year 2000

```python
cf = bf[bf['release_year'] > 1999]
ff = bf[bf['release_year'] < 1999]
```

FINDING SUM OF BOTH

```python
a_sum = cf['count'].sum()
b_sum = ff['count'].sum()
```

Making a list of labels and values, for plotting a piew chart

```python
values = [a_sum,b_sum]
labels = ['After 2000','Before 2000']
```

MAKing the pie chart

```python
plt.pie(values, labels=labels,autopct='%1.1f%%')
plt.title = 'movies released by year'
plt.show()

```

DISTRIBUTION OF RATINGS


```python
df = pd.read_csv('C:/Users/lijaz/Desktop/DATA/netflix/netflix1.csv')

bf = df.groupby(['rating']).size().reset_index(name='count')

values = bf['count']
labels = bf['rating']

plt.figure(figsize=(10,8))
plt.pie(values,labels=labels)
plt.legend(loc='lower left',labels=labels)

plt.show()

```
